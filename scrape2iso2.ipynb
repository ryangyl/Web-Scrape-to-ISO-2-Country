```python
new_data = {}  
compiled = {}
for company in companies:
    url = "https://google.com/search?q=where+is+" + company
    r = requests.get(url, verify=False)
    soup = BeautifulSoup(r.text, "lxml")
    element = soup.find_all('div')
    for e in element:
        content_div = e.get_text(strip=True)
        x = find_countries(content_div)
        if x:  
            country_object = x[0][0]
            alpha_2_value = country_object.alpha_2  
            new_data = {company: alpha_2_value}  
            print(new_data)
            break  
    compiled.update(new_data)
print(compiled)
```

```python
import pandas as pd

df = pd.DataFrame(data=compiled, index=[0])

df = (df.T)
with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also

    display(df)
```
